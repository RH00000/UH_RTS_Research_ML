```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('gridsearch_same_confidence_with_3_ways_to_skip.csv')

# Ensure correct types
df['shared_thr'] = df['shared_thr'].astype(float)
df['skip_b_thresh'] = df['skip_b_thresh'].astype(float)
df['skip_c_thresh'] = df['skip_c_thresh'].astype(float)
```

```python
# Use np.isclose to avoid float comparison issues
df_plot = df[np.isclose(df['shared_thr'], 0.85)]

# Check if df_plot is empty
print(f"Rows matching shared_thr=0.85: {len(df_plot)}")
print(df_plot.head())

# Create 3D plot
fig = px.scatter_3d(df_plot,
                    x='skip_b_thresh',
                    y='skip_c_thresh',
                    z='avg_time',
                    color='avg_time',
                    title='Avg Time vs Skip Thresholds (shared_thr=0.85)',
                    labels={'skip_b_thresh': 'Skip B', 'skip_c_thresh': 'Skip C', 'avg_time': 'Time'},
                    width=1000,
                    height=700)

fig.show()
```

```python
fig.write_html("plot_3d_avg_time_vs_skip_thresh_shared_conf_0.85.html")
```
